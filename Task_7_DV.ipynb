```python
# ---------- Step 1: Import Libraries ----------
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from wordcloud import WordCloud
import networkx as nx
import matplotlib.pyplot as plt

# ---------- Step 2: Download NLTK Resources ----------
nltk.download('punkt_experimental')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')


# ---------- Step 3: Input Text Dataset ----------
text = """
Artificial Intelligence (AI) is transforming the way we live and work.
AI systems are now capable of analyzing large amounts of data, recognizing patterns,
and making decisions faster than humans. From healthcare to finance, AI-driven applications
are improving efficiency and accuracy. However, ethical challenges such as data privacy,
bias, and transparency remain critical concerns. Continuous research and innovation
are helping to build trustworthy AI systems that can benefit society as a whole.
"""

# ---------- Step 4: Text Preprocessing ----------
tokens = word_tokenize(text.lower())
filtered = [word for word in tokens if word.isalpha() and word not in stopwords.words('english')]

# ---------- Step 5: Generate Word Cloud ----------
wordcloud = WordCloud(width=900, height=450, background_color='white', colormap='plasma').generate(' '.join(filtered))

# ---------- Step 6: Display Word Cloud ----------
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - Artificial Intelligence Text", fontsize=16)
plt.show()

# ---------- Step 7: Create Text Network ----------
sentences = sent_tokenize(text.lower())
G = nx.Graph()

for sentence in sentences:
    words = [word for word in word_tokenize(sentence) if word.isalpha() and word not in stopwords.words('english')]
    for i in range(len(words)):
        for j in range(i + 1, len(words)):
            if G.has_edge(words[i], words[j]):
                G[words[i]][words[j]]['weight'] += 1
            else:
                G.add_edge(words[i], words[j], weight=1)

# ---------- Step 8: Analyze Central Nodes ----------
degree_centrality = nx.degree_centrality(G)
sorted_nodes = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("\nTop 5 Central Words in the Network:")
for word, centrality in sorted_nodes[:5]:
    print(f"{word}: {centrality:.4f}")

# ---------- Step 9: Visualize Text Network ----------
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.3, seed=42)  # Force-based layout (FruchtermanReingold)
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=800, alpha=0.7)
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
plt.title("Text Network Visualization - Word Co-occurrence Graph", fontsize=16)
plt.axis('off')
plt.show()
```

```python
# ---------- Step 1: Import Libraries ----------
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from wordcloud import WordCloud
import networkx as nx
import matplotlib.pyplot as plt

# ---------- Step 2: Download NLTK Resources ----------
nltk.download('punkt')
nltk.download('stopwords')

# ---------- Step 3: Input Text Dataset ----------
text = """
Climate change refers to long-term alterations in temperature, precipitation, wind patterns,
and other elements of the Earth's climate system. The main driver of recent climate change
is human activity, particularly the burning of fossil fuels such as coal, oil, and gas.
This releases greenhouse gases like carbon dioxide and methane into the atmosphere,
causing global warming. Rising temperatures lead to melting ice caps, rising sea levels,
and more frequent extreme weather events such as floods and droughts.
Addressing climate change requires international cooperation, renewable energy adoption,
and sustainable development practices to protect our planet for future generations.
"""

# ---------- Step 4: Text Preprocessing ----------
tokens = word_tokenize(text.lower())
filtered = [word for word in tokens if word.isalpha() and word not in stopwords.words('english')]

# ---------- Step 5: Generate Word Cloud ----------
wordcloud = WordCloud(
    width=900,
    height=450,
    background_color='white',
    colormap='viridis'
).generate(' '.join(filtered))

# ---------- Step 6: Display Word Cloud ----------
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud - Climate Change Text", fontsize=16)
plt.show()

# ---------- Step 7: Create Text Network ----------
sentences = sent_tokenize(text.lower())
G = nx.Graph()

for sentence in sentences:
    words = [word for word in word_tokenize(sentence) if word.isalpha() and word not in stopwords.words('english')]
    for i in range(len(words)):
        for j in range(i + 1, len(words)):
            if G.has_edge(words[i], words[j]):
                G[words[i]][words[j]]['weight'] += 1
            else:
                G.add_edge(words[i], words[j], weight=1)

# ---------- Step 8: Analyze Central Nodes ----------
degree_centrality = nx.degree_centrality(G)
sorted_nodes = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
print("\nTop 5 Central Words in the Climate Change Network:")
for word, centrality in sorted_nodes[:5]:
    print(f"{word}: {centrality:.4f}")

# ---------- Step 9: Visualize Text Network ----------
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.3, seed=42)  # Force-based layout
nx.draw_networkx_nodes(G, pos, node_color='lightgreen', node_size=900, alpha=0.8)
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.4)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
plt.title("Text Network Visualization - Word Co-occurrence Graph (Climate Change)", fontsize=16)
plt.axis('off')
plt.show()
```
