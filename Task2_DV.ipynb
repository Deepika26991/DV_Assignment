```python
# ===============================
# Exploratory Data Analysis (EDA) - Fitness Tracking
# ===============================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load dataset
df = pd.read_csv("fitness_tracking.csv")

# Step 2: Convert WorkoutDate to datetime
df['WorkoutDate'] = pd.to_datetime(df['WorkoutDate'])

# Set Seaborn style
sns.set(style="whitegrid")

# ===============================
#  CATEGORICAL DATA VISUALIZATION
# ===============================

# Bar Chart - Frequency of each WorkoutType
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='WorkoutType', palette='pastel')
plt.title("Bar Chart: Frequency of Workout Types")
plt.show()

# Pie Chart - Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart: Gender Distribution")
plt.show()

# ===============================
#  CONTINUOUS DATA VISUALIZATION
# ===============================

# Histogram - Calories Burned
plt.figure(figsize=(6,4))
sns.histplot(df['CaloriesBurned'], kde=False, color='skyblue')
plt.title("Histogram: Calories Burned")
plt.show()

# Density Plot - Duration Minutes
plt.figure(figsize=(6,4))
sns.kdeplot(df['DurationMinutes'], shade=True, color='orange')
plt.title("Density Plot: Workout Duration")
plt.show()

# Rug Plot - Calories Burned
plt.figure(figsize=(6,4))
sns.rugplot(df['CaloriesBurned'], height=0.05)
plt.title("Rug Plot: Calories Burned")
plt.show()

# Strip Plot - Calories Burned
plt.figure(figsize=(6,4))
sns.stripplot(x=df['CaloriesBurned'], color='green')
plt.title("Strip Plot: Calories Burned")
plt.show()

# Swarm Plot - Calories Burned by WorkoutType
plt.figure(figsize=(8,4))
sns.swarmplot(x='WorkoutType', y='CaloriesBurned', data=df)
plt.title("Swarm Plot: Calories Burned by Workout Type")
plt.show()

# Scatter Plot - Duration vs CaloriesBurned
plt.figure(figsize=(6,4))
sns.scatterplot(x='DurationMinutes', y='CaloriesBurned', data=df, hue='WorkoutType', palette='Set2', s=100)
plt.title("Scatter Plot: Duration vs Calories Burned")
plt.show()

# Line Plot - Total Calories Burned over WorkoutDate
df_grouped = df.groupby('WorkoutDate')['CaloriesBurned'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.lineplot(data=df_grouped, x='WorkoutDate', y='CaloriesBurned', marker='o')
plt.title("Line Plot: Total Calories Burned Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
